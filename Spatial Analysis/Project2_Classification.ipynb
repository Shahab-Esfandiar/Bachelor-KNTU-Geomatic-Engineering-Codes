### Practice 1

classify the following list of nominal scale

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
nominal = np.chararray((8,8))

for i in range(len(nominal)):
    
    for j in range(len(nominal)):
        
        nominal[i,j] = chr(np.random.randint(ord('m'), ord('x'))) # Create random raster
        
print("\nUnclassified raster :\n")         
print(nominal)
print('----------------------------------------------------------')

nominal_classified = np.zeros((8,8), int)

l = np.unique(nominal) # Create classes with values
c = [0] * 4
c[0] = l[0:1]
c[1] = l[1:4]
c[2] = l[4:8]
c[3] = l[8:11]

print("\nUnique values of raster :\n")
print(l)
print('----------------------------------------------------------')

print("\nOur classes with Manual interval method :\n") 

for x in c:
    print(x, '\n')
    
print('----------------------------------------------------------')

for (i, j), z in np.ndenumerate(nominal): # Classification loop

    for k in range(len(c)):
        
        if z in c[k]:    
           nominal_classified[i,j] = k+1

print("\nClassified raster :\n")
print(nominal_classified)
print('----------------------------------------------------------\n')

arr_LV = nominal_classified # Plot result
fig, ax = plt.subplots()
ax.matshow(arr_LV)
ax.set_title('After Classification')

for (i, j), z in np.ndenumerate(arr_LV):
    ax.text(j, i, '{}'.format(z), ha='center', va='center', size=12)
    
```

### Practice 2

classify following raster with use of equal interval method, you should implement the method from scratch

start value for calasses should be zero and written code should handle random rasters like the one below

```python
arr_c1 = np.random.randint(19, 34, size=(7, 7)) # Create random raster

print("\nUnclassified raster :\n")         
print(arr_c1)
print('--------------------------------------------------------')
 
arr_c1_classified = arr_c1.copy()
l=np.unique(arr_c1)

print("\nUnique values of raster :\n")
print(l)
print('--------------------------------------------------------')

n = 5 
c = math.ceil((l[-1]-l[0])/n)
classes = [l[i:i + c] for i in range(0, len(l), c)] # Compute classes's length and create them with values

print("\nOur classes with Equal interval method :\n")

for x in classes:
    print(x, '\n')
    
print('--------------------------------------------------------')
    
for (i, j), z in np.ndenumerate(arr_c1): # Classification loop 
    
    for k in range(n):
        
        if z in classes[k]:
           arr_c1_classified[i,j] = k
                
print("\nClassified raster :\n")               
print(arr_c1_classified)
print('\n------------------------------------------------------------------------------------------------------------\n')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10)) # Plot result
ax1.matshow(arr_c1)
ax1.set_title('Before Classification')
arr_c1f = arr_c1_classified
ax2.matshow(arr_c1f)
ax2.set_title('After Classification')

for (i, j), z in np.ndenumerate(arr_c1):
    ax1.text(j, i, '{}'.format(z), ha='center', va='center', size=12)

for (i, j), z in np.ndenumerate(arr_c1f):
    ax2.text(j, i, '{}'.format(z), ha='center', va='center', size=12)
```

```python
arr_c2 = np.random.randint(512, 2156, size=(7, 7)) # Create random raster

print("\nUnclassified raster :\n")         
print(arr_c2)
print('------------------------------------------------------------------------')
 
arr_c2_classified = arr_c2.copy()
l=np.unique(arr_c2)

print("\nUnique values of raster :\n")
print(l)
print('------------------------------------------------------------------------')

n = 3
c = math.pow((l[-1]/l[0]), 1/n) # Compute classes's length and create them with values
classes = [0] * n

print("\nOur classes with Equal interval method :\n")

for x in range(n):
    classes[x] = [l for l in range((math.ceil(l[0]* math.pow(c, x))), (math.ceil(l[0]* math.pow(c, x+1)))+1)]
    classes[x] = list(set(classes[x]) & set(l))
    classes[x].sort()
    print(classes[x], '\n')
    
print('------------------------------------------------------------------------')

for (i, j), z in np.ndenumerate(arr_c2): # Classification loop
            
    for k in range(n):
        
        if z in classes[k]:
           arr_c2_classified[i,j] = k+1 
                
print("\nClassified raster :\n")                
print(arr_c2_classified)
print('\n------------------------------------------------------------------------------------------------------------\n')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10)) # Plot result
ax1.matshow(arr_c2)
ax1.set_title('Before Classification')
arr_c2f = arr_c2_classified
ax2.matshow(arr_c2f)
ax2.set_title('After Classification')

for (i, j), z in np.ndenumerate(arr_c2):
    ax1.text(j, i, '{}'.format(z), ha='center', va='center', size=12)

for (i, j), z in np.ndenumerate(arr_c2f):
    ax2.text(j, i, '{}'.format(z), ha='center', va='center', size=12)
```

### Practice 4

classify the folloing raster according to given table


```python
arr_c3 = np.random.randint(1, 20, size=(6, 6)) 
arr_c3[np.random.randint(len(arr_c3)),1], arr_c3[np.random.randint(len(arr_c3)),0], arr_c3[np.random.randint(len(arr_c3)),3] = [0,100,-25]

print("\nUnclassified raster :\n")         
print(arr_c3)
print('------------------------------------------------------------------------')

arr_c3_classified = arr_c3.copy() # Create random raster
l=np.unique(arr_c3)

print("\nUnique values of raster :\n")
print(l)
print('------------------------------------------------------------------------') # Create classes with values

c = [0] * 9
c[0] = [1,2,3]  
c[1] = [4,5,6,7]
c[2] = [8]
c[3] = [9,10,11,12]
c[4] = [13,14,15]
c[5] = [16]
c[6] = [17,18,19]
c[7] = [20]
c[8] = [-25,0,100]
    
for (i, j), z in np.ndenumerate(arr_c3): # Classification loop
        
    if z in c[0] or z in c[3] or z in c[6]:
       arr_c3_classified[i,j] = 5

    elif z in c[1]:
       arr_c3_classified[i,j] = 3
    
    elif z in c[2] or z in c[8]:
       arr_c3_classified[i,j] = 1
               
    elif z in c[4]:
       arr_c3_classified[i,j] = 2   
               
    else:
       arr_c3_classified[i,j] = 4             
            
print("\nClassified raster with Define interval method :\n")                
print(arr_c3_classified)
print('\n------------------------------------------------------------------------------------------------------------\n')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 10)) # Plot result
ax1.matshow(arr_c3)
ax1.set_title('Before Classification')
arr_c3f = arr_c3_classified
ax2.matshow(arr_c3f)
ax2.set_title('After Classification')

for (i, j), z in np.ndenumerate(arr_c3):
    ax1.text(j, i, '{}'.format(z), ha='center', va='center', size=12)

for (i, j), z in np.ndenumerate(arr_c3f):
    ax2.text(j, i, '{}'.format(z), ha='center', va='center', size=12)
```
